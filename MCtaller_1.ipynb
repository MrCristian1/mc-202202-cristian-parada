```python
from ast import Num


A = set()
datos = int(input("Escriba la cantidad de datos que quiera tener en su primer conjunto"))
for i in range(1,datos+1):
    Num = float(input("Escriba el numero decimal: "))
    A.add(Num)
print("Su primer conjunto es:", A)

F = set()
datos = int(input("Escriba la cantidad de datos que quiera tener en su segundo conjunto"))
for i in range(1,datos+1):
    Num = float(input("Escriba el numero decimal: "))
    F.add(Num)
print("Su segundo conjunto es:", F)

def pedirNumeroEntero():
 
    correcto=False
    num=0
    while(not correcto):
        try:
            num = int(input("Introduce un numero entero: "))
            correcto=True
        except ValueError:
            print('Error, introduce un numero entero')
     
    return num
 
salir = False
opcion = 0
 
while not salir:
    print("Seleccione una de las 4 operaciones que desea hacer con sus 2 conjuntos")
    print("¿Que desea hacer?")
 
    print ("1. Union")
    print ("2. Intersección")
    print ("3. Diferencia")
    print ("4. Diferencia simétrica")
    print ("5. Salir")
     
    print ("Elige una opcion")
 
    opcion = pedirNumeroEntero()
    U = set()
    I = set()
    D = set()
    Dsim = set()
    if opcion == 1:
      for i in A:
        U.add(i)
      for i in F:
        U.add(i)
      print("La union da como resultado:", U)
    elif opcion == 2:
      for i in A:
        if i in F:
          I.add(i)
      print("La intersección da como resultado:", I)
    elif opcion == 3:
      for i in A:
        if i in F:
          pass
        else:
          D.add(i)
      print("La diferencia da como resultado:", D)
    elif opcion == 4:
      for i in A:
        if i in F:
          pass
        else:
          Dsim.add(i)
      for i in F:
        if i in A:
          pass
        else:
          Dsim.add(i)
      print("La diferencia simétrica da como resultado:", Dsim)

    elif opcion == 5:
        salir = True
    else:
        print ("Introduce un numero entre 1 y 5")
 
print ("Fin")
```
